```python
data = pd.read_csv('Shopping_data.csv')
data.head(10)
```

```python
data.drop(columns=['CustomerID','Genre'],axis=1)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

data.plot(kind='scatter',x='Age',y='Spending Score (1-100)',color='blue')
data.plot(kind='scatter',x='Annual Income (k$)',y='Spending Score (1-100)',color='green')
```

```python
feature_cols = ['Age', 'Spending Score (1-100)']
X = data[feature_cols]
X.head()
```

```python
y =  data['Annual Income (k$)']
y.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1)
```

```python
print(X_train.shape)
print(y_train.shape)
print(X_test.shape)
print(y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression
linreg = LinearRegression()
model = linreg.fit(X_train, y_train)
```

```python
y_pred = linreg.predict(X_test)
print(y_pred)
```
